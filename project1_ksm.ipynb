# 문제1-7) '지역고용통계' 를 위한 전처리 함수를 작성하세요.

````python
# 데이터를 전처리 하기 위해서는 수치화와 누락데이터 처리가 중요하다.
# '지역고용통계' 를 위해 수치화해야 하는 데이터는 '지역','전체종사자수', '빈일자리수', '입직자수', '이직자수' 이다
# 원본의 메타정보상 '전체종사자수' 와 '빈일자리수' 는 4월 말 기준이고, '입직자수', '이직자수' 는 4월 중 기준이므로
# '전체 종사자수' + '빈 일자리수' = '일자리 총합' 으로 정의한다.
# '지역'은 '시도' 를 기준으로 나눈다.
def goyong_preprocessing(path):
    raw_goyong = pd.read_excel(path, sheet_name = '데이터', skiprows=1)
    select_col = raw_goyong.loc[:, ['지역별','전체종사자','빈일자리','입직자','이직자',]]
    select_col.columns =['지역명','고용자수','빈일자리','입직자수','이직자수']
    goyong_data = select_col
    goyong_data['고용자수'] = pd.to_numeric(goyong_data['고용자수'])
    goyong_data['빈일자리'] = pd.to_numeric(goyong_data['빈일자리'])
    goyong_data['입직자수'] = pd.to_numeric(goyong_data['입직자수'])
    goyong_data['일자리 총합'] = goyong_data['고용자수'] + goyong_data['빈일자리']
    
# 지역명이라는 기준으로 대단위로 묶인 데이터들 보기좋게 정리하기.
# 지역명 컬럼의 칸이 비어있지 않으면 내용 그대로, 비어있다면 한칸 전의 개체 출력하도록 함수 지정    
    big_col = []
    for num,temp in enumerate(goyong_data['지역명']):
        if pd.isna(temp) :
            big_col.append(big_col[num-1])
        else:
            big_col.append(temp)
# 인덱스 정렬
    goyong_data.index = [big_col]
    
    def goyong(x):
        temp=x.split(' ')[0]
        if len(temp) != 4 :
            return temp[:2]
        else:
            return temp[0] + temp[2]
        
    goyong_data['시도'] = goyong_data['지역명'].dropna().apply(goyong)
    goyong_data['구군'] = goyong_data['지역명'].dropna().apply(lambda x: x.split(' ')[1])
    
    return goyong_data
````

# 문제1-8) 전처리 함수를 사용해서 전처리된 결과를 출력하세요.

````python
goyong_df=goyong_preprocessing(goyong_path)
goyong_df
````

````python
# 전처리한 결과에서 '시도' 를 index로 시도별 합계를 구한다.
goyong_sido = goyong_df.groupby('시도').sum()
# 일자리 총합의 값을 기준으로 내림차순으로 정렬한다.
goyong_sido.sort_values(by='일자리 총합', ascending=False)
````

# 문제2-5) '주민등록세대수' 전처리 함수 만들기

````python
def juminpreprocessing(path):
    raw_jumin = pd.read_excel(path, sheet_name = '데이터')
    raw_jumin = select_col
    select_col.columns = ['시도','구군','세대수']
    
    big_col = []
    for num,temp in enumerate(raw_jumin['시도']):
        if pd.isna(temp) :
            big_col.append(big_col[num-1])
        else:
            big_col.append(temp)
                    
    def jumin(x):
        temp=x.split(' ')[0]
        if len(temp) != 4 :
            return temp[:2]
        else:
            return temp[0] + temp[2]
        
        raw_jumin['시도'] = big_col
        raw_jumin['시도'] = raw_jumin['시도'].dropna().apply(jumin)        
        jumin_data = raw_jumin[raw_jumin['구군'] != '소계']
        
    return jumin_data    
````

# 문제2-7) 시도 단위로 고용자수 출력

````python
# 시도별 '전산업' 종사자 출력. 후 '고용자수' 를 기준으로 내림차순 정렬
jiyuk_sido = jiyuk_data.groupby('시도').sum()
jiyuk_sido.sort_values(by='고용자수', ascending=False)
````

````python
# '세대수' 와 '고용자수' 를 한눈에 보고 비교하고 싶어서 합쳐봄
# '세대수' 를 앞에 표시하고 싶어서 순서를 바꿔봄.
gy_sido= jiyuk_sido + jumin_sido 
gy_sido['고용자수'] = jiyuk_sido['고용자수']
gy_sido['세대수'] = jumin_sido['세대수']
gy_sido = gy_sido[['세대수', '고용자수']]
gy_sido
````
