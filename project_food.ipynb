```python
def food(x):
    food = pd.concat([food_2008[food_2008['2008_식품명'] == x],food_2009[food_2009['2009_식품명'] == x],food_2010[food_2010['2010_식품명'] == x], food_2011[food_2011['2011_식품명'] == x], food_2012[food_2012['2012_식품명'] == x], food_2013[food_2013['2013_식품명'] == x], food_2014[food_2014['2014_식품명'] == x], food_2015[food_2015['2015_식품명'] == x], food_2016[food_2016['2016_식품명'] == x], food_2017[food_2017['2017_식품명'] == x]], sort=False)
    return food
```

```python
배추김치 = food('배추김치').copy()
잡곡밥 = food('잡곡밥').copy()
쌀밥 = food('쌀밥').copy()
커피 = food('커피').copy()
우유 = food('우유').copy()
된장국 = food('된장국').copy()
멸치볶음 = food('멸치볶음').copy()
된장찌개 = food('된장찌개').copy()
```

```python
def year(x):
    big_col = []
    for num, temp in enumerate(x['2008_식품명']):
        if pd.isna(temp) :
            big_col.append(big_col[num-1])
        else:
            big_col.append(temp)
    x['2008_식품명'] = big_col
    x.index = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
    x= x.drop(['2009_식품명','2010_식품명','2011_식품명','2012_식품명','2013_식품명','2014_식품명','2015_식품명','2016_식품명','2017_식품명'],axis =1).copy()
    x.columns = ['식품명','빈도(%)']
    return x
```

```python
kimchi = year(배추김치).copy()
japgok = year(잡곡밥).copy()
ssal = year(쌀밥).copy()
coffee = year(커피).copy()
milk = year(우유).copy()
djgook = year(된장국).copy()
chovy = year(멸치볶음).copy()
dj = year(된장찌개).copy()
```

```python
plt.figure(figsize=(15,8))
col_kimchi = kimchi['빈도(%)'].copy()
col_japgok = japgok['빈도(%)'].copy()
col_ssal = ssal['빈도(%)'].copy()
col_coffee = coffee['빈도(%)'].copy()
col_milk = milk['빈도(%)'].copy()
col_djgook = djgook['빈도(%)'].copy()
col_chovy = chovy['빈도(%)'].copy()
col_dj = dj['빈도(%)'].copy()
plt.plot(col_kimchi, label ='배추김치')
plt.plot(col_japgok, label = '잡곡밥')
plt.plot(col_ssal, label ='쌀밥')
plt.plot(col_milk, label = '우유')
plt.plot(col_coffee, label = '커피')
plt.plot(col_djgook, label = '된장국')
plt.plot(col_chovy, label = '멸치볶음')
plt.plot(col_dj , label = '된장찌개')
plt.title('식습관 변화(2008~2017)')
plt.legend(loc='upper left')
plt.show()
```

## 1. 커피의 섭취 빈도가 2008년에는 잡곡밥, 쌀밥 보다 낮았으나, 점점 증가하여 2017년에는 커피의 섭취 빈도가 잡곡밥, 쌀밥 보다 높아졌다.
## 2. 웰빙의 영향으로 잡곡밥의 섭취 빈도가 쌀밥보다 훨씬 높아졌을거라고 예상했지만 , 2017년에는 쌀밥이 섭취 빈도가 더 높아졌다.
